```python
import pandas as pd
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 5 zestawie≈Ñ i 5 wykresow
df = pd.read_csv('weekly-gasoline.csv')
df.drop('Fiscal Week', axis=1, inplace=True)
df['Fiscal Year'] = pd.to_datetime(df['Fiscal Year'], format='%Y')
```

```python
df.info()
```

```python
df_year =df.groupby('Fiscal Year').mean()

x=df_year.index
y=df_year["Current Year Production"]
z = df_year["Previous Year Production"]
```

```python
df_year.head(4)
```

```python
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(x, y, "r.--", label="≈örednia produkcja benzyny")
ax.plot(x, z, "b.--", label="≈örednia produkcja benzyny z poprzedniego roku")
plt.title("≈örednia produkcja benzyny w USA")
ax.legend()
```
